### [Github](https://github.com/)

#### Setting up an [SSH](https://en.wikipedia.org/wiki/Secure_Shell) passkey

[SSH Keys and Github.mp4](https://www.youtube.com/watch?v=aHcflUMfCp8)

***Note\*: The procedure is similar for all OS's (Windows/Linux/Mac)***

#### 0. Create a [github](https://github.com/) profile

#### 1. download [gitbash](https://git-scm.com/downloads)

Install gitbash on your local machine ...

Add the following to your environment path variable in windows `C:\Program Files\Git\usr\bin` to have a "Linux-like" experience with the terminal - this enables you to use built-in's like `nano` etc ...

#### 2. Run the following commands in the terminal:

`> ssh-keygen -t rsa -C "name-of-your-device"`

By accepting the default location for the ssh key file, the information will be stored in `C:\Users\<user_name>\`

["rsa"](https://en.wikipedia.org/wiki/RSA_(cryptosystem)) is an encryption algorithm designed by (Rivest, Shamir & Adleman), others include [EdDSA/ed25519](https://en.wikipedia.org/wiki/EdDSA) and more ...

You can also specify your own path for storing the ssh key file, for example on the desktop: `C:\Users\<user_name>\Desktop\.ssh`

Next, you should fill in a passphrase, which is optional, but a good security measure ... 

This will produce the following output in the terminal ...

... the following files are produced ...

The contents of the `id_rsa.pub` file looks like this ...

It contains a ***public*** key which can be sent over a(n) (insecure) internet connection ...

### 3. Set up SSH on Github Account/Profile

LOG into your github profile, navigate to "settings" then "SSH and GPG keys" then click on "new SSH key" to create a new SSH key, provide an identifiable title and paste in the contents of the id_rsa.pub file into the text-area "key" box, then click on "Add SSH key"

If you did not choose the default file name for your ssh key files, then you will have to do an additional step of notifying the ssh agent where to find the files - this is easily achieved by creating a `config` file in the `~/.ssh` folder and pasting the following contents in it ...

`IdentityFile /path/to/ssh/key/file`

#### 4. Now run the following `test` to confirm that you have access to your github account via your local machine ...

`> ssh -T git@github.com`

You will get the following message:

`ECDSA key fingerprint is SHA256:p2QAMXNIC1TJYWeIOttrVc98/R1BUFWu3/LiyKgUfQM.
Are you sure you want to continue connecting (yes/no/[fingerprint])?`

Type "yes" and press "Enter"

You should see the following response ...

`Hi <username>! You've successfully authenticated, but GitHub does not provide shell access.`

#### 5. Create a Repository on Github

LOG into your github profile and navigate to "repositories" then click on "create a new repository" - Don't close the window (the window contains instructions on how to upload your project)

If you accidentally closed the window, and you are not sure of what you are doing, it's best to go to the repository settings and then delete the repository all together and start from scratch ...

Once you have created an empty repository, you can work from your local machine to upload your first code onto github ...
